**Above dataset has 5 columns in which only one column has numerical values so will move on using that column for now.**

```python
df = pd.read_csv("../input/swiggy-thane-restaurants-data/Swiggy_thane_restaurants_data.csv")
fig = df['For Two'].plot()
```

**After plotting the data as it is, the graph seems unpredictable. So, lets print it by sorting the data.**

```python
sorted_df = df. sort_values(by=["For Two"], ascending=True)
sorted_df.head()
```

**So, above is the sorted dataset. Lets plot it and see it that gives any value to the analysis.**

```python
fig = sorted_df['For Two'].plot()
```

**The above graph is more complex than the earlier one. So, lets use Seaborn library of Python which is a data visualization library based on matplotlib. And it provides a high-level interface for drawing attractive and informative statistical graphics.**

```python
sns.set(rc={'figure.figsize':(10, 8)})
sns.distplot(df["For Two"])
```

**Finally, this graph shows that maximum of the restaurants from Thane which are listed in Swiggy serves for two person in less than Rs.500.**

```python
mean = df['For Two'].mean()
sum = df['For Two'].sum()
max = df['For Two'].max()
min = df['For Two'].min()
count = df['For Two'].count()
median = df['For Two'].median() 
std = df['For Two'].std() 
var = df['For Two'].var()  
groupby_count = df.groupby(['For Two']).count()

print ('Mean value: ' + str(mean))
print ('Sum of values: ' + str(sum))
print ('Max value: ' + str(max))
print ('Min value: ' + str(min))
print ('Count of values: ' + str(count))
print ('Median value: ' + str(median))
print ('Std of values: ' + str(std))
print ('Var of values: ' + str(var))
print ('Count of values, grouped by the range: ' + str(groupby_count))
```

**More precisely it is for 200 and 300.**

```python
#finding the percentage of restaurants which serves for specific amount.
((df['For Two'].value_counts())/count)*100
```

**Approximately 50% of the total restaurants registered with Swiggy in Thane are serving two person in 200 and 300.**
